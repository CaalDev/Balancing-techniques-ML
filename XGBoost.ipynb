```python
project_base_dir = r"/content/drive/Shareddrives/PROYECTOS/1"
file = 'newdatos-Junio14-dataset.csv'
file_path = os.path.join(project_base_dir, file)

df = pd.read_csv(file_path, sep=';')

conditions = [
    df['Tratamiento'] == 'Control',
    df['Tratamiento'] == 'E_Hidrico'
]
choices = [0, 1]
default_value = 2
df['Tratamiento'] = np.select(conditions, choices, default=default_value)
df['Tratamiento'] = df['Tratamiento'].astype(int)

X = df.iloc[:, 3:].values
y = df['Tratamiento'].values

y_encoded = y
descriptive_target_names = ['Control (0)', 'E_Hidrico (1)', 'Otras (2)']
numerical_labels = [0, 1, 2]
scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)


drive_base_path = '/content/drive/Shareddrives/PROYECTOS/1'
output_spectral_data_path = os.path.join(drive_base_path, 'data', 'spectral_data')
output_spectral_labels_path = os.path.join(drive_base_path, 'data', 'spectral_labels')

os.makedirs(output_spectral_data_path, exist_ok=True)
os.makedirs(output_spectral_labels_path, exist_ok=True)
np.savetxt(os.path.join(output_spectral_data_path, '0_spectral_data.txt'), X_train)
np.savetxt(os.path.join(output_spectral_labels_path, '0_spectral_labels.txt'), y_train, fmt='%d') # Usar %d para enteros

print(f"- {os.path.join(output_spectral_data_path, '0_spectral_data.txt')}")
print(f"- {os.path.join(output_spectral_labels_path, '0_spectral_labels.txt')}")
```

```python
t3 = time.time()

##############################################################################
"""args for AE"""

args = {}
args['dim_h'] = 256 # Dimensión de la capa oculta intermedia del AE
args['input_dim'] = 2151 # Dimensión de tus datos espectrales (2500 - 350 + 1 = 2151)
args['n_z'] = 128 # Dimensión del espacio latente (código)
args['sigma'] = 1.0 # Parámetro para la pérdida de diversidad (mse2)
args['lambda_'] = 0.01 # Ponderación para la pérdida de diversidad (mse2)
args['lr'] = 0.0002 # Learning rate
args['epochs'] = 200 # Número de épocas de entrenamiento
args['batch_size'] = 100 # Tamaño del batch
args['save'] = True # Guardar el modelo
args['train'] = True # Bandera para entrenar

##############################################################################

## create encoder model and decoder model
class Encoder(nn.Module):
    def __init__(self, args):
        super(Encoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.input_dim, self.dim_h * 4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 4, self.dim_h * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 2, self.n_z)
        )
    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

class Decoder(nn.Module):
    def __init__(self, args):
        super(Decoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.n_z, self.dim_h * 2),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 2, self.dim_h * 4),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 4, self.input_dim),
            nn.Tanh() # CONFIRMADO: Coincide con la normalización de datos a [-1, 1]
        )
    def forward(self, x):
        x = self.fc(x)
        return x

##############################################################################
"""set models, loss functions"""
def free_params(module: nn.Module):
    for p in module.parameters():
        p.requires_grad = True

def frozen_params(module: nn.Module):
    for p in module.parameters():
        p.requires_grad = False

##############################################################################
"""functions to create SMOTE images (estas funciones son auxiliares, no se usan directamente en el loop de entrenamiento del AE)"""
def biased_get_class(c):
    # dec_x y dec_y deben ser accesibles globalmente o pasados como argumentos
    # Esta función se usará más adelante en la generación de muestras, no en el entrenamiento del AE.
    xbeg = dec_x[dec_y == c]
    ybeg = dec_y[dec_y == c]
    return xbeg, ybeg

def G_SM(X, y, n_to_sample, cl):
    # Esta función es la lógica SMOTE original, no se usa directamente para entrenar el AE.
    n_neigh = 5 + 1
    nn = NearestNeighbors(n_neighbors=n_neigh, n_jobs=1)
    nn.fit(X)
    dist, ind = nn.kneighbors(X)
    base_indices = np.random.choice(list(range(len(X))), n_to_sample, replace=True)
    neighbor_indices = np.random.choice(list(range(1, n_neigh)), n_to_sample)
    X_base = X[base_indices]
    X_neighbor = X[ind[base_indices, neighbor_indices]]
    samples = X_base + np.multiply(np.random.rand(n_to_sample, 1),
                                   X_neighbor - X_base)
    return samples, [cl] * n_to_sample

##############################################################################

# --- RUTAS DE ARCHIVOS (AJUSTADAS) ---
# La carpeta base donde se guardaron los archivos .txt en el paso anterior.
# ESTO ES CRÍTICO: DEBE COINCIDIR CON LA SALIDA DEL SCRIPT DE PREPARACIÓN.
drive_base_for_data = '/content/drive/Shareddrives/PROYECTOS/1'

# Rutas completas a las carpetas donde están los archivos .txt
dtrnimg_path = os.path.join(drive_base_for_data, 'data', 'spectral_data')
dtrnlab_path = os.path.join(drive_base_for_data, 'data', 'spectral_labels')

# Ruta donde se guardarán los modelos entrenados. (Puede ser diferente a dtrnimg_path)
project_base_dir_for_models = '/content/drive/Shareddrives/PROYECTOS/1'
modpth = os.path.join(project_base_dir_for_models, 'models_DeepSMOTE')
os.makedirs(modpth, exist_ok=True) # Crea la carpeta 'models_DeepSMOTE' si no existe

# Obtener la lista de archivos .txt
ids_img = [f for f in os.listdir(dtrnimg_path) if f.endswith('.txt') and 'spectral_data' in f]
ids_lab = [f for f in os.listdir(dtrnlab_path) if f.endswith('.txt') and 'spectral_labels' in f]

idtri_f = [os.path.join(dtrnimg_path, image_id) for image_id in ids_img]
idtrl_f = [os.path.join(dtrnlab_path, image_id) for image_id in ids_lab]

print("Archivos de características (spectral_data.txt) encontrados:", idtri_f)
print("Archivos de etiquetas (spectral_labels.txt) encontrados:", idtrl_f)


# Loop para procesar cada conjunto de datos (normalmente habrá solo uno: '0_spectral_data.txt')
for i in range(len(idtri_f)): # O range(1) si estás seguro de que solo tienes un archivo
    print(f"\nProcesando archivo {i}")
    encoder = Encoder(args)
    decoder = Decoder(args)

    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Usando dispositivo: {device}")
    decoder = decoder.to(device)
    encoder = encoder.to(device)

    criterion = nn.MSELoss()
    criterion = criterion.to(device)

    trnimgfile = idtri_f[i]
    trnlabfile = idtrl_f[i]

    print(f"Cargando {trnimgfile}")
    print(f"Cargando {trnlabfile}")

    dec_x = np.loadtxt(trnimgfile)
    dec_y = np.loadtxt(trnlabfile) # Las etiquetas deben ser dtype=int64 para Counter y operaciones de clase

    print('Shape de datos espectrales de entrenamiento cargados: ', dec_x.shape)
    print('Shape de etiquetas de entrenamiento cargadas: ', dec_y.shape)
    print("Conteo de clases en el dataset de entrenamiento cargado:", collections.Counter(dec_y))

    batch_size = args['batch_size']
    num_workers = 0 # Mantener en 0 si tienes problemas con CUDA y multiprocessing

    tensor_x = torch.Tensor(dec_x)
    tensor_y = torch.tensor(dec_y, dtype=torch.long) # Asegurar dtype long para etiquetas

    dataset_bal = TensorDataset(tensor_x, tensor_y)
    train_loader = torch.utils.data.DataLoader(dataset_bal,
                                               batch_size=batch_size, shuffle=True, num_workers=num_workers)

    best_loss = np.inf

    t0 = time.time()
    if args['train']:
        enc_optim = torch.optim.Adam(encoder.parameters(), lr = args['lr'])
        dec_optim = torch.optim.Adam(decoder.parameters(), lr = args['lr'])

        for epoch in range(args['epochs']):
            train_loss = 0.0
            tmse_loss = 0.0 # Loss de reconstrucción (mse)
            tdiscr_loss = 0.0 # Loss de diversidad (mse2)

            encoder.train()
            decoder.train()

            for data, labels in train_loader:
                encoder.zero_grad()
                decoder.zero_grad()

                data, labels = data.to(device), labels.to(device)

                z_hat = encoder(data) # Codifica los datos
                x_hat = decoder(z_hat) # Decodifica el espacio latente

                mse = criterion(x_hat, data) # Pérdida de reconstrucción (MSE)

                # --- Cálculo de la pérdida de diversidad (mse2) ---
                # Esto es crucial para DeepSMOTE: busca forzar la diversidad en el espacio latente
                # seleccionando una clase minoritaria al azar y penalizando la reconstrucción
                # de muestras cercanas a ella en el espacio latente.

                # Obtener conteos de clase del dataset COMPLETO de entrenamiento (dec_y)
                # Esto asegura que las clases minoritarias se identifiquen correctamente.
                global_class_counts = collections.Counter(dec_y)
                majority_count = max(global_class_counts.values())
                minority_classes = [label for label, count in global_class_counts.items() if count < majority_count]

                if not minority_classes: # Si no hay clases minoritarias (o todas son iguales), no hay pérdida de diversidad
                    mse2 = torch.tensor(0.0).to(device)
                else:
                    # Seleccionar una clase minoritaria presente en el batch actual, si es posible
                    available_minority_in_batch = [cl for cl in minority_classes if cl in labels.unique().cpu().numpy()]
                    if available_minority_in_batch:
                        tc = np.random.choice(available_minority_in_batch, 1)[0]
                    else: # Si ninguna minoritaria está en el batch, elige una al azar del batch
                        if len(labels.unique()) > 0:
                            tc = np.random.choice(labels.unique().cpu().numpy(), 1)[0]
                        else: # Si el batch está vacío (caso extremo)
                            mse2 = torch.tensor(0.0).to(device)
                            tc = None

                    if tc is not None:
                        # Obtener todas las muestras de la clase 'tc' del dataset COMPLETO
                        xbeg_spectral = dec_x[dec_y == tc]
                        if xbeg_spectral.shape[0] > 1: # Necesita al menos 2 muestras para el "shift"
                            xbeg_tensor = torch.Tensor(xbeg_spectral).to(device)
                            z_class = encoder(xbeg_tensor).detach().cpu().numpy() # Latent codes of this class
                            zclen = len(z_class)
                            if zclen > 1:
                                zcminus = np.arange(1, zclen) # Indices 1, 2, ..., N-1
                                zcplus = np.append(zcminus, 0) # Indices 1, 2, ..., N-1, 0 (cyclic shift)
                                zc_enc = z_class[zcplus, :] # Shifted latent codes
                                zc_enc_tensor = torch.Tensor(zc_enc).to(device)
                                ximg_generated_shifted = decoder(zc_enc_tensor) # Reconstrucción de códigos latentes "shifted"
                                xcnew_spectral = xbeg_spectral[zcplus, :] # Datos originales "shifted"
                                xcnew_spectral_tensor = torch.Tensor(xcnew_spectral).to(device)
                                # MSE entre la reconstrucción de los códigos shifted y los datos originales shifted
                                # Esto es una forma de pérdida de diversidad: fuerza al AE a no reconstruir "demasiado bien"
                                # un dato si su representación latente es ligeramente alterada de su vecino.
                                mse2 = criterion(ximg_generated_shifted, xcnew_spectral_tensor)
                            else: # No suficientes muestras para el shift
                                mse2 = torch.tensor(0.0).to(device)
                        else: # No suficientes muestras en la clase
                            mse2 = torch.tensor(0.0).to(device)
                    else: # Si tc es None
                        mse2 = torch.tensor(0.0).to(device)


                # Combinación de pérdidas: reconstrucción (mse) + diversidad (mse2)
                # El parámetro lambda_ pondera la importancia de la pérdida de diversidad.
                comb_loss = mse + args['lambda_'] * mse2
                comb_loss.backward()

                enc_optim.step()
                dec_optim.step()

                train_loss += comb_loss.item() * data.size(0)
                tmse_loss += mse.item() * data.size(0)
                tdiscr_loss += mse2.item() * data.size(0)

            train_loss = train_loss / len(train_loader.dataset)
            tmse_loss = tmse_loss / len(train_loader.dataset)
            tdiscr_loss = tdiscr_loss / len(train_loader.dataset)

            print('Epoch: {} \tTrain Loss: {:.6f} \tmse loss: {:.6f} \tmse2 loss: {:.6f}'.format(epoch,
                                                                                                 train_loss, tmse_loss, tdiscr_loss))

            if train_loss < best_loss:
                print('Guardando modelo \'best\'...')
                os.makedirs(os.path.join(modpth, str(i)), exist_ok=True)
                path_bst_enc = os.path.join(modpth, str(i), 'bst_enc.pth')
                path_bst_dec = os.path.join(modpth, str(i), 'bst_dec.pth')
                torch.save(encoder.state_dict(), path_bst_enc)
                torch.save(decoder.state_dict(), path_bst_dec)
                best_loss = train_loss

        # Guardar el modelo final (siempre se guarda)
        os.makedirs(os.path.join(modpth, str(i)), exist_ok=True)
        path_f_enc = os.path.join(modpth, str(i), 'f_enc.pth')
        path_f_dec = os.path.join(modpth, str(i), 'f_dec.pth')
        print(f"Guardando modelos finales: {path_f_enc}, {path_f_dec}")
        torch.save(encoder.state_dict(), path_f_enc)
        torch.save(decoder.state_dict(), path_f_dec)
        print()

    t1 = time.time()
    print('Tiempo total de entrenamiento (min): {:.2f}'.format((t1 - t0)/60))

t4 = time.time()
print('Tiempo final de ejecución del script (min): {:.2f}'.format((t4 - t3)/60))
```

```python
# Suprimir advertencias (especialmente de use_label_encoder en XGBoost)
warnings.filterwarnings("ignore")

project_base_dir = '/content/drive/Shareddrives/PROYECTOS/1/'
csv_file_name = 'newdatos-Junio14-dataset.csv'
csv_file_path = os.path.join(project_base_dir, csv_file_name)

deepsmote_data_temp_dir = project_base_dir
deepsmote_models_dir = os.path.join(project_base_dir, 'models_DeepSMOTE')
deepsmote_balanced_data_dir = os.path.join(project_base_dir, 'Balanced_Spectral_Data')
os.makedirs(deepsmote_models_dir, exist_ok=True)
os.makedirs(deepsmote_balanced_data_dir, exist_ok=True)

df = pd.read_csv(csv_file_path, sep=';')

conditions = [
    df['Tratamiento'] == 'Control',
    df['Tratamiento'] == 'E_Hidrico'
]
choices = [0, 1]
default_value = 2
df['Tratamiento'] = np.select(conditions, choices, default=default_value)
df['Tratamiento'] = df['Tratamiento'].astype(int)
X = df.iloc[:, 3:].values
y_encoded = df['Tratamiento'].values

scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled = scaler.fit_transform(X)
X_train_scaled, X_test_scaled, y_train_encoded, y_test_encoded = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

spectral_data_filename = os.path.join(deepsmote_data_temp_dir, '0_spectral_data.txt')
spectral_labels_filename = os.path.join(deepsmote_data_temp_dir, '0_spectral_labels.txt')

np.savetxt(spectral_data_filename, X_train_scaled)
np.savetxt(spectral_labels_filename, y_train_encoded)

print(f"\nDatos de entrenamiento para DeepSMOTE guardados en: {spectral_data_filename} y {spectral_labels_filename}")

# --- 4. DeepSMOTE: Entrenamiento del Autoencoder ---
print("\n--- DeepSMOTE: Entrenamiento del Autoencoder ---")
print(f"CUDA disponible: {torch.cuda.is_available()}")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Usando dispositivo: {device}")

# Args para el Autoencoder (hiperparámetros)
args = {
    'dim_h': 256,
    'input_dim': X_train_scaled.shape[1], # Debe ser la cantidad de columnas de tus características
    'n_z': 128, # Dimensión del espacio latente
    'sigma': 1.0,
    'lambda': 0.01,
    'lr': 0.0002,
    'epochs': 200,
    'batch_size': 100,
    'save': True,
    'train': True
}

# Definición de Encoder y Decoder (arquitectura de la red neuronal)
class Encoder(nn.Module):
    def __init__(self, args):
        super(Encoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.input_dim, self.dim_h * 4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 4, self.dim_h * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 2, self.n_z)
        )
    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

class Decoder(nn.Module):
    def __init__(self, args):
        super(Decoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.n_z, self.dim_h * 2),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 2, self.dim_h * 4),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 4, self.input_dim),
            nn.Tanh() # La función de activación Tanh es adecuada si los datos están escalados a [-1, 1]
        )
    def forward(self, x):
        x = self.fc(x)
        return x

# Instanciar modelos y cargar datos de entrenamiento para el Autoencoder
encoder = Encoder(args).to(device)
decoder = Decoder(args).to(device)
criterion = nn.MSELoss().to(device) # Criterio de pérdida: Mean Squared Error
enc_optim = torch.optim.Adam(encoder.parameters(), lr=args['lr']) # Optimizador Adam para el Encoder
dec_optim = torch.optim.Adam(decoder.parameters(), lr=args['lr']) # Optimizador Adam para el Decoder

dec_x_train = np.loadtxt(spectral_data_filename)
dec_y_train = np.loadtxt(spectral_labels_filename).astype(int) # Asegurar tipo int

tensor_x_train = torch.Tensor(dec_x_train)
tensor_y_train = torch.tensor(dec_y_train, dtype=torch.long)
dataset_ae = TensorDataset(tensor_x_train, tensor_y_train)
train_loader = DataLoader(dataset_ae, batch_size=args['batch_size'], shuffle=True, num_workers=0)

best_loss = np.inf
t0_ae_train = time.time()

print("Iniciando entrenamiento del Autoencoder...")
for epoch in range(args['epochs']):
    encoder.train()
    decoder.train()
    train_loss = 0.0
    tmse_loss = 0.0
    tdiscr_loss = 0.0

    for data, labels in train_loader:
        encoder.zero_grad()
        decoder.zero_grad()

        data, labels = data.to(device), labels.to(device)

        z_hat = encoder(data)
        x_hat = decoder(z_hat)
        mse = criterion(x_hat, data) # Pérdida de reconstrucción

        # Lógica para la pérdida de discriminación de DeepSMOTE (similar a SMOTE pero en espacio latente)
        global_class_counts = collections.Counter(dec_y_train)
        majority_count = max(global_class_counts.values())
        minority_classes = [label for label, count in global_class_counts.items() if count < majority_count]

        mse2 = torch.tensor(0.0).to(device) # Inicializar a 0.0
        if minority_classes:
            available_minority_in_batch = [cl for cl in minority_classes if cl in labels.unique().cpu().numpy()]
            if available_minority_in_batch:
                tc = np.random.choice(available_minority_in_batch, 1)[0]
                xbeg_spectral = dec_x_train[dec_y_train == tc]
                if xbeg_spectral.shape[0] > 1: # Asegurarse de que haya al menos 2 muestras para el "shift"
                    xbeg_tensor = torch.Tensor(xbeg_spectral).to(device)
                    z_class = encoder(xbeg_tensor).detach().cpu().numpy()
                    if len(z_class) > 1:
                        zcminus = np.arange(1, len(z_class))
                        zcplus = np.append(zcminus, 0)
                        zc_enc = z_class[zcplus, :] # Desplaza las muestras en el espacio latente
                        zc_enc_tensor = torch.Tensor(zc_enc).to(device)
                        ximg_generated_shifted = decoder(zc_enc_tensor) # Reconstruye desde el espacio latente desplazado
                        xcnew_spectral = xbeg_spectral[zcplus, :] # Las muestras originales desplazadas
                        xcnew_spectral_tensor = torch.Tensor(xcnew_spectral).to(device)
                        mse2 = criterion(ximg_generated_shifted, xcnew_spectral_tensor) # Pérdida de discriminación

        comb_loss = mse + args['lambda'] * mse2 # Pérdida combinada
        comb_loss.backward() # Retropropagación

        enc_optim.step() # Actualizar pesos del Encoder
        dec_optim.step() # Actualizar pesos del Decoder

        train_loss += comb_loss.item() * data.size(0)
        tmse_loss += mse.item() * data.size(0)
        tdiscr_loss += mse2.item() * data.size(0)

    train_loss /= len(train_loader.dataset)
    tmse_loss /= len(train_loader.dataset)
    tdiscr_loss /= len(train_loader.dataset)

    print('Epoch: {} \tTrain Loss: {:.6f} \tmse loss: {:.6f} \tmse2 loss: {:.6f}'.format(
        epoch, train_loss, tmse_loss, tdiscr_loss))

    # Guardar el mejor modelo basado en la pérdida de entrenamiento
    if train_loss < best_loss:
        path_bst_enc = os.path.join(deepsmote_models_dir, 'bst_enc.pth')
        path_bst_dec = os.path.join(deepsmote_models_dir, 'bst_dec.pth')
        torch.save(encoder.state_dict(), path_bst_enc)
        torch.save(decoder.state_dict(), path_bst_dec)
        best_loss = train_loss

path_f_enc = os.path.join(deepsmote_models_dir, 'f_enc.pth')
path_f_dec = os.path.join(deepsmote_models_dir, 'f_dec.pth')
torch.save(encoder.state_dict(), path_f_enc)
torch.save(decoder.state_dict(), path_f_dec)
print(f"Modelos finales de DeepSMOTE guardados en: {path_f_enc} y {path_f_dec}")

t1_ae_train = time.time()
print('Tiempo total de entrenamiento del Autoencoder (min): {:.2f}'.format((t1_ae_train - t0_ae_train)/60))


print("\n--- DeepSMOTE: Generación de Muestras Sintéticas ---")
# Función G_SM para generar muestras en el espacio latente (basada en SMOTE)
def G_SM(X_latent, n_to_sample, cl):
    n_neigh = 5 + 1 # 5 vecinos + la propia muestra
    nn = NearestNeighbors(n_neighbors=n_neigh, n_jobs=1)
    nn.fit(X_latent)
    dist, ind = nn.kneighbors(X_latent)
    base_indices = np.random.choice(list(range(len(X_latent))), n_to_sample, replace=True)
    neighbor_indices = np.random.choice(list(range(1, n_neigh)), n_to_sample) # Excluye el propio punto
    X_base = X_latent[base_indices]
    X_neighbor = X_latent[ind[base_indices, neighbor_indices]]
    # Interpolar entre una muestra y su vecino en el espacio latente
    samples = X_base + np.multiply(np.random.rand(n_to_sample, 1), X_neighbor - X_base)
    return samples, [cl] * n_to_sample

# Cargar los modelos entrenados para inferencia
encoder_gen = Encoder(args).to(device)
decoder_gen = Decoder(args).to(device)

encoder_gen.load_state_dict(torch.load(path_f_enc, map_location=device))
decoder_gen.load_state_dict(torch.load(path_f_dec, map_location=device))

encoder_gen.eval() # Poner el modelo en modo evaluación
decoder_gen.eval()

# Convertir el conjunto de entrenamiento original (escalado, codificado) a tensor
tensor_x_train_scaled = torch.Tensor(X_train_scaled).to(device)

with torch.no_grad(): # Desactivar el cálculo de gradientes para la inferencia
    z_enc_all = encoder_gen(tensor_x_train_scaled).cpu().numpy() # Obtener representaciones latentes

combx = X_train_scaled.copy()
comby = y_train_encoded.copy()

# Determinar las clases minoritarias a sobremuestrear
class_counts_train = collections.Counter(y_train_encoded)
actual_majority_count = max(class_counts_train.values())
minority_classes_encoded = [label for label, count in class_counts_train.items() if count < actual_majority_count]

print(f"Conteo de clases en el conjunto de entrenamiento antes de balancear: {class_counts_train}")
print(f"Clases minoritarias (codificadas) a sobremuestrear: {minority_classes_encoded}")
print(f"Conteo de la clase mayoritaria: {actual_majority_count}")

for minority_class_label_encoded in minority_classes_encoded:
    z_class_minority = z_enc_all[y_train_encoded == minority_class_label_encoded]
    num_minority_samples = len(z_class_minority)

    n_to_generate = actual_majority_count - num_minority_samples

    if n_to_generate > 0:
        print(f"Generando {n_to_generate} muestras para la clase minoritaria codificada: {minority_class_label_encoded}")
        # Generar muestras sintéticas en el espacio latente
        smote_z_samples, smote_y_labels_encoded = G_SM(z_class_minority, n_to_generate, minority_class_label_encoded)
        # Decodificar las muestras del espacio latente de vuelta al espacio original de las características
        smote_x_decoded = decoder_gen(torch.Tensor(smote_z_samples).to(device)).detach().cpu().numpy()

        combx = np.concatenate((combx, smote_x_decoded), axis=0)
        comby = np.concatenate((comby, np.array(smote_y_labels_encoded)), axis=0)
    else:
        print(f"La clase codificada {minority_class_label_encoded} ya tiene suficientes o más muestras ({num_minority_samples}) que la clase mayoritaria ({actual_majority_count}). No se generarán muestras.")

X_train_balanced = combx
y_train_balanced = comby.astype(int) # Asegurarse de que las etiquetas sigan siendo int

print(f"\nShape final del dataset de entrenamiento balanceado (características): {X_train_balanced.shape}")
print(f"Shape final del dataset de entrenamiento balanceado (etiquetas): {y_train_balanced.shape}")
print(f"Conteo final de clases en el dataset de entrenamiento balanceado: {collections.Counter(y_train_balanced)}")

# Guardar los datos balanceados (opcional, pero útil para reusar)
balanced_data_file = os.path.join(deepsmote_balanced_data_dir, '0_spectral_balanced_data.txt')
balanced_labels_file = os.path.join(deepsmote_balanced_data_dir, '0_spectral_balanced_labels.txt')
np.savetxt(balanced_data_file, X_train_balanced)
np.savetxt(balanced_labels_file, y_train_balanced)
print(f"Datos balanceados guardados en: {balanced_data_file} y {balanced_labels_file}")


# --- 6. Entrenamiento del Modelo XGBoost ---
print("\n--- Entrenamiento del Modelo XGBoost ---")

num_unique_classes_final = len(np.unique(y_train_balanced))

xgb_model = xgb.XGBClassifier(
    objective='multi:softmax', # Para clasificación multiclase
    num_class=num_unique_classes_final, # Número de clases de salida
    eval_metric='mlogloss', # Métrica de evaluación para clasificación multiclase
    n_estimators=100, # Número de árboles en el bosque
    learning_rate=0.1,
    max_depth=6,
    use_label_encoder=False, # Importante para versiones recientes de XGBoost
    random_state=42,
    n_jobs=-1 # Usar todos los núcleos disponibles de la CPU
)

xgb_model.fit(X_train_balanced, y_train_balanced) # Entrenar con los datos balanceados por DeepSMOTE
print("Entrenamiento del XGBoost completado.")


# --- 7. Evaluación del Modelo XGBoost ---
print("\n--- Evaluación del Modelo XGBoost en el Conjunto de Prueba Original ---")
y_pred_xgb = xgb_model.predict(X_test_scaled) # Predecir sobre el conjunto de prueba original

accuracy_xgb = accuracy_score(y_test_encoded, y_pred_xgb)
print(f"Precisión (Accuracy) del modelo XGBoost: {accuracy_xgb:.4f}")

# Nombres descriptivos para el reporte de clasificación y la matriz de confusión
# Estos nombres corresponden al mapeo 0='Control', 1='E_Hidrico', 2='Otras' ('Sana')
descriptive_target_names = ['Control (0)', 'E_Hidrico (1)', 'Otras (2)']

print("\nReporte de Clasificación:")
print(classification_report(y_test_encoded, y_pred_xgb, target_names=descriptive_target_names))

conf_matrix_xgb = confusion_matrix(y_test_encoded, y_pred_xgb)
print("\nMatriz de Confusión:")
print(conf_matrix_xgb)


# --- 8. Graficar la Matriz de Confusión ---
print("\n--- Graficando la Matriz de Confusión ---")

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_xgb, annot=True, fmt='d', cmap='viridis', cbar=False,
            xticklabels=descriptive_target_names, yticklabels=descriptive_target_names)

plt.title('Matriz de Confusión de XGBoost (DeepSMOTE Balanced Data)')
plt.xlabel('Clase Predicha')
plt.ylabel('Clase Real')
plt.show()

print("\n--- Proceso Completo Finalizado ---")
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import collections
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset
from sklearn.neighbors import NearestNeighbors
import time
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suprimir advertencias (especialmente de use_label_encoder en XGBoost)
warnings.filterwarnings("ignore")

print("--- Inicio del Proceso Completo: DeepSMOTE + XGBoost ---")

# --- Configuración de Rutas ---
project_base_dir = '/content/drive/Shareddrives/PROYECTOS/1/'
csv_file_name = 'newdatos-Junio14-dataset.csv'
csv_file_path = os.path.join(project_base_dir, csv_file_name)

# Directorios para archivos temporales y modelos de DeepSMOTE
deepsmote_data_temp_dir = project_base_dir # Usamos la carpeta principal para los .txt temporales
deepsmote_models_dir = os.path.join(project_base_dir, 'models_DeepSMOTE')
deepsmote_balanced_data_dir = os.path.join(project_base_dir, 'Balanced_Spectral_Data')

os.makedirs(deepsmote_models_dir, exist_ok=True)
os.makedirs(deepsmote_balanced_data_dir, exist_ok=True)

print(f"\nArchivos temporales DeepSMOTE se guardarán en: {deepsmote_data_temp_dir}")
print(f"Modelos DeepSMOTE se guardarán en: {deepsmote_models_dir}")
print(f"Datos balanceados DeepSMOTE se guardarán en: {deepsmote_balanced_data_dir}")

# --- 1. Carga del Dataset Original y Preprocesamiento ---
print(f"\nCargando dataset original desde: {csv_file_path}")
try:
    df = pd.read_csv(csv_file_path, sep=';')
except FileNotFoundError:
    print(f"\nERROR: No se encontró el archivo {csv_file_name}.")
    print(f"Asegúrate de que el archivo exista en la ruta: {csv_file_path}")
    exit()

print(f"Dimensiones del DataFrame original: {df.shape}")
print("Columnas del DataFrame:", df.columns.tolist())

spectral_columns = [str(i) for i in range(350, 2501)]
existing_spectral_columns = [col for col in spectral_columns if col in df.columns]
X = df[existing_spectral_columns].values
y = df['Sana'].values

print(f"Conteo de clases en etiquetas ORIGINALES (sin codificar): {collections.Counter(y)}")

# Codificar etiquetas (-1, 0, 1 a 0, 1, 2)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
original_class_labels = le.classes_ # Guardar las etiquetas originales para la decodificación

print(f"\nMapeo de etiquetas (Original -> Codificado):")
for orig_lbl, enc_lbl in zip(sorted(np.unique(y)), sorted(np.unique(y_encoded))):
    print(f"  {orig_lbl} -> {enc_lbl}")
print(f"Conteo de clases en etiquetas CODIFICADAS: {collections.Counter(y_encoded)}")

# Escalar los datos (MinMaxScaler para DeepSMOTE)
scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled = scaler.fit_transform(X)
print(f"Dimensiones de X (escalado): {X_scaled.shape}")

# --- 2. División de Datos (Entrenamiento y Prueba) ---
# Usamos y_encoded para asegurar que las etiquetas en los splits sean 0, 1, 2
X_train_scaled, X_test_scaled, y_train_encoded, y_test_encoded = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\nDimensiones X_train (escalado): {X_train_scaled.shape}")
print(f"Dimensiones y_train (codificado): {y_train_encoded.shape}")
print(f"Conteo de clases en y_train (codificado, antes de DeepSMOTE): {collections.Counter(y_train_encoded)}")
print(f"Dimensiones X_test (escalado): {X_test_scaled.shape}")
print(f"Dimensiones y_test (codificado): {y_test_encoded.shape}")
print(f"Conteo de clases en y_test (codificado): {collections.Counter(y_test_encoded)}")

# --- 3. Preparación de Archivos .txt para DeepSMOTE ---
spectral_data_filename = os.path.join(deepsmote_data_temp_dir, '0_spectral_data.txt')
spectral_labels_filename = os.path.join(deepsmote_data_temp_dir, '0_spectral_labels.txt')

np.savetxt(spectral_data_filename, X_train_scaled)
np.savetxt(spectral_labels_filename, y_train_encoded)

print(f"\nDatos de entrenamiento para DeepSMOTE guardados en: {spectral_data_filename} y {spectral_labels_filename}")

# --- 4. DeepSMOTE: Entrenamiento del Autoencoder ---
print("\n--- DeepSMOTE: Entrenamiento del Autoencoder ---")
print(f"CUDA disponible: {torch.cuda.is_available()}")
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Usando dispositivo: {device}")

# Args para el Autoencoder
args = {
    'dim_h': 256,
    'input_dim': X_train_scaled.shape[1], # Debe ser 2151
    'n_z': 128,
    'sigma': 1.0,
    'lambda': 0.01,
    'lr': 0.0002,
    'epochs': 200,
    'batch_size': 100,
    'save': True,
    'train': True
}

# Definición de Encoder y Decoder
class Encoder(nn.Module):
    def __init__(self, args):
        super(Encoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.input_dim, self.dim_h * 4),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 4, self.dim_h * 2),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(self.dim_h * 2, self.n_z)
        )
    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

class Decoder(nn.Module):
    def __init__(self, args):
        super(Decoder, self).__init__()
        self.input_dim = args['input_dim']
        self.dim_h = args['dim_h']
        self.n_z = args['n_z']
        self.fc = nn.Sequential(
            nn.Linear(self.n_z, self.dim_h * 2),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 2, self.dim_h * 4),
            nn.ReLU(True),
            nn.Linear(self.dim_h * 4, self.input_dim),
            nn.Tanh() # Debe coincidir con la normalización a [-1, 1]
        )
    def forward(self, x):
        x = self.fc(x)
        return x

# Instanciar modelos y cargar datos
encoder = Encoder(args).to(device)
decoder = Decoder(args).to(device)
criterion = nn.MSELoss().to(device)
enc_optim = torch.optim.Adam(encoder.parameters(), lr=args['lr'])
dec_optim = torch.optim.Adam(decoder.parameters(), lr=args['lr'])

dec_x_train = np.loadtxt(spectral_data_filename)
dec_y_train = np.loadtxt(spectral_labels_filename).astype(int) # Asegurar tipo int

tensor_x_train = torch.Tensor(dec_x_train)
tensor_y_train = torch.tensor(dec_y_train, dtype=torch.long)
dataset_bal = TensorDataset(tensor_x_train, tensor_y_train)
train_loader = torch.utils.data.DataLoader(dataset_bal, batch_size=args['batch_size'], shuffle=True, num_workers=0)

best_loss = np.inf
t0_ae_train = time.time()

print("Iniciando entrenamiento del Autoencoder...")
for epoch in range(args['epochs']):
    encoder.train()
    decoder.train()
    train_loss = 0.0
    tmse_loss = 0.0
    tdiscr_loss = 0.0

    for data, labels in train_loader:
        encoder.zero_grad()
        decoder.zero_grad()

        data, labels = data.to(device), labels.to(device)

        z_hat = encoder(data)
        x_hat = decoder(z_hat)
        mse = criterion(x_hat, data)

        global_class_counts = collections.Counter(dec_y_train)
        majority_count = max(global_class_counts.values())
        minority_classes = [label for label, count in global_class_counts.items() if count < majority_count]

        mse2 = torch.tensor(0.0).to(device) # Default a 0 si no hay minorías o no se encuentra clase
        if minority_classes:
            available_minority_in_batch = [cl for cl in minority_classes if cl in labels.unique().cpu().numpy()]
            if available_minority_in_batch:
                tc = np.random.choice(available_minority_in_batch, 1)[0]
                xbeg_spectral = dec_x_train[dec_y_train == tc]
                if xbeg_spectral.shape[0] > 1: # Asegurarse de que haya al menos 2 muestras para el "shift"
                    xbeg_tensor = torch.Tensor(xbeg_spectral).to(device)
                    z_class = encoder(xbeg_tensor).detach().cpu().numpy()
                    if len(z_class) > 1:
                        zcminus = np.arange(1, len(z_class))
                        zcplus = np.append(zcminus, 0)
                        zc_enc = z_class[zcplus, :]
                        zc_enc_tensor = torch.Tensor(zc_enc).to(device)
                        ximg_generated_shifted = decoder(zc_enc_tensor)
                        xcnew_spectral = xbeg_spectral[zcplus, :]
                        xcnew_spectral_tensor = torch.Tensor(xcnew_spectral).to(device)
                        mse2 = criterion(ximg_generated_shifted, xcnew_spectral_tensor)

        comb_loss = mse + mse2
        comb_loss.backward()

        enc_optim.step()
        dec_optim.step()

        train_loss += comb_loss.item() * data.size(0)
        tmse_loss += mse.item() * data.size(0)
        tdiscr_loss += mse2.item() * data.size(0)

    train_loss /= len(train_loader.dataset)
    tmse_loss /= len(train_loader.dataset)
    tdiscr_loss /= len(train_loader.dataset)

    print('Epoch: {} \tTrain Loss: {:.6f} \tmse loss: {:.6f} \tmse2 loss: {:.6f}'.format(
        epoch, train_loss, tmse_loss, tdiscr_loss))

    if train_loss < best_loss:
        # print('Guardando modelo \'best\'...') # No imprimir cada vez, solo si hay cambio
        path_bst_enc = os.path.join(deepsmote_models_dir, 'bst_enc.pth')
        path_bst_dec = os.path.join(deepsmote_models_dir, 'bst_dec.pth')
        torch.save(encoder.state_dict(), path_bst_enc)
        torch.save(decoder.state_dict(), path_bst_dec)
        best_loss = train_loss

# Guardar el modelo final (siempre se guarda)
path_f_enc = os.path.join(deepsmote_models_dir, 'f_enc.pth')
path_f_dec = os.path.join(deepsmote_models_dir, 'f_dec.pth')
torch.save(encoder.state_dict(), path_f_enc)
torch.save(decoder.state_dict(), path_f_dec)
print(f"Modelos finales de DeepSMOTE guardados en: {path_f_enc} y {path_f_dec}")

t1_ae_train = time.time()
print('Tiempo total de entrenamiento del Autoencoder (min): {:.2f}'.format((t1_ae_train - t0_ae_train)/60))


# --- 5. DeepSMOTE: Generación de Muestras Sintéticas ---
print("\n--- DeepSMOTE: Generación de Muestras Sintéticas ---")
# Función G_SM para generar muestras en el espacio latente
def G_SM(X_latent, n_to_sample, cl):
    n_neigh = 5 + 1 # 5 vecinos + la propia muestra
    nn = NearestNeighbors(n_neighbors=n_neigh, n_jobs=1)
    nn.fit(X_latent)
    dist, ind = nn.kneighbors(X_latent)
    base_indices = np.random.choice(list(range(len(X_latent))), n_to_sample, replace=True)
    neighbor_indices = np.random.choice(list(range(1, n_neigh)), n_to_sample) # Excluye el propio punto
    X_base = X_latent[base_indices]
    X_neighbor = X_latent[ind[base_indices, neighbor_indices]]
    samples = X_base + np.multiply(np.random.rand(n_to_sample, 1), X_neighbor - X_base)
    return samples, [cl] * n_to_sample

# Cargar los modelos entrenados para inferencia
encoder_gen = Encoder(args).to(device)
decoder_gen = Decoder(args).to(device)

encoder_gen.load_state_dict(torch.load(path_f_enc, map_location=device))
decoder_gen.load_state_dict(torch.load(path_f_dec, map_location=device))

encoder_gen.eval()
decoder_gen.eval()

# Convertir el conjunto de entrenamiento original (escalado, codificado) a tensor
tensor_x_train_scaled = torch.Tensor(X_train_scaled).to(device)

with torch.no_grad():
    z_enc_all = encoder_gen(tensor_x_train_scaled).cpu().numpy()

combx = X_train_scaled.copy()
comby = y_train_encoded.copy()

# Determinar las clases minoritarias basadas en las etiquetas codificadas
class_counts_train = collections.Counter(y_train_encoded)
actual_majority_count = max(class_counts_train.values())
minority_classes_encoded = [label for label, count in class_counts_train.items() if count < actual_majority_count]

print(f"Conteo de clases en el conjunto de entrenamiento antes de balancear: {class_counts_train}")
print(f"Clases minoritarias (codificadas) a sobremuestrear: {minority_classes_encoded}")
print(f"Conteo de la clase mayoritaria: {actual_majority_count}")

for minority_class_label_encoded in minority_classes_encoded:
    z_class_minority = z_enc_all[y_train_encoded == minority_class_label_encoded]
    num_minority_samples = len(z_class_minority)

    n_to_generate = actual_majority_count - num_minority_samples

    if n_to_generate > 0:
        print(f"Generando {n_to_generate} muestras para la clase minoritaria codificada: {minority_class_label_encoded}")
        # Generar en el espacio latente
        smote_z_samples, smote_y_labels_encoded = G_SM(z_class_minority, n_to_generate, minority_class_label_encoded)
        # Decodificar de vuelta al espacio original
        smote_x_decoded = decoder_gen(torch.Tensor(smote_z_samples).to(device)).detach().cpu().numpy()

        combx = np.concatenate((combx, smote_x_decoded), axis=0)
        comby = np.concatenate((comby, np.array(smote_y_labels_encoded)), axis=0)
    else:
        print(f"La clase codificada {minority_class_label_encoded} ya tiene suficientes o más muestras ({num_minority_samples}) que la clase mayoritaria ({actual_majority_count}). No se generarán muestras.")

X_train_balanced = combx
y_train_balanced = comby.astype(int) # Asegurarse de que las etiquetas sigan siendo int

print(f"\nShape final del dataset de entrenamiento balanceado (características): {X_train_balanced.shape}")
print(f"Shape final del dataset de entrenamiento balanceado (etiquetas): {y_train_balanced.shape}")
print(f"Conteo final de clases en el dataset de entrenamiento balanceado: {collections.Counter(y_train_balanced)}")

# Guardar los datos balanceados
balanced_data_file = os.path.join(deepsmote_balanced_data_dir, '0_spectral_balanced_data.txt')
balanced_labels_file = os.path.join(deepsmote_balanced_data_dir, '0_spectral_balanced_labels.txt')
np.savetxt(balanced_data_file, X_train_balanced)
np.savetxt(balanced_labels_file, y_train_balanced)
print(f"Datos balanceados guardados en: {balanced_data_file} y {balanced_labels_file}")


# --- 6. Entrenamiento del Modelo XGBoost ---
print("\n--- Entrenamiento del Modelo XGBoost ---")
import xgboost as xgb

num_unique_classes_final = len(np.unique(y_train_balanced))

xgb_model = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=num_unique_classes_final,
    eval_metric='mlogloss',
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    use_label_encoder=False, # Importante para versiones recientes de XGBoost
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(X_train_balanced, y_train_balanced)
print("Entrenamiento del XGBoost completado.")


# --- 7. Evaluación del Modelo XGBoost ---
print("\n--- Evaluación del Modelo XGBoost en el Conjunto de Prueba Original ---")
y_pred_xgb = xgb_model.predict(X_test_scaled)

accuracy_xgb = accuracy_score(y_test_encoded, y_pred_xgb)
print(f"Precisión (Accuracy) del modelo XGBoost: {accuracy_xgb:.4f}")

# Nombres descriptivos para el reporte de clasificación y la matriz de confusión
# Usamos inverse_transform del LabelEncoder para obtener los nombres originales
descriptive_target_names = [str(label) for label in le.inverse_transform(np.arange(num_unique_classes_final))]

print("\nReporte de Clasificación:")
print(classification_report(y_test_encoded, y_pred_xgb, target_names=descriptive_target_names))

conf_matrix_xgb = confusion_matrix(y_test_encoded, y_pred_xgb)
print("\nMatriz de Confusión:")
print(conf_matrix_xgb)


# --- 8. Graficar la Matriz de Confusión ---
print("\n--- Graficando la Matriz de Confusión ---")

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_xgb, annot=True, fmt='d', cmap='viridis', cbar=False,
            xticklabels=descriptive_target_names, yticklabels=descriptive_target_names)

plt.title('Matriz de Confusión de XGBoost (DeepSMOTE Balanced Data)')
plt.xlabel('Clase Predicha')
plt.ylabel('Clase Real')
plt.show()

print("\n--- Proceso Completo Finalizado ---")
```
